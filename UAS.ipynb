```python
df = data.copy()
#ubah nilai sc_w dan px_height yang 0 menjadi nilai mean dari masing2 kolom trsebut
avg_sc_w = df['sc_w'].mean()
avg_px_height = df['px_height'].mean()


df['sc_w'].replace({0:avg_sc_w}, inplace=True)
df['px_height'].replace({0:avg_px_height}, inplace=True)
```

### Screen Size

Konversi screen_size dari cm ke inches (data sc_h dan sc_w)

```python
# Buat variable/kolom baru dengan nama sc_size

import numpy as np

df['sc_size'] = np.sqrt((df['sc_h']**2) + (df['sc_w']**2))
df['sc_size'] = round(df['sc_size']/2.54, 2)
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
sns.kdeplot(data=df, x='sc_size', hue='price_range', ax=axs[0])
sns.boxplot(data=df, x='price_range', y='sc_size', ax=axs[1])
plt.show()
```

Ukuran Layar menunjukkan sedikit variasi sepanjang variabel target. Hal ini dapat membantu dalam memprediksi kategori sasaran.

```python
# hapus kolom sc_h dan sc_w karena sudah menggunakan kolom screen_size
df.drop(['sc_h', 'sc_w'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['price_range'], axis = 1)
y = df['price_range']
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
# Scaling values of X

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python

# Split data menjadi train and test sets

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.20, random_state = 42)
     
```

```python
X_train.shape
```

```python
y_train.shape
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python

from sklearn.metrics import classification_report

# Prediction

y_pred_test = lr.predict(X_test)
y_pred_train = lr.predict(X_train)
print('Classification report for Logistic Regression (Test set)= ')
print(classification_report(y_pred_test, y_test))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

#Generate the confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred_test)

print(cf_matrix)

ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels([0,1,2,3])
ax.yaxis.set_ticklabels([0,1,2,3])

## Display the visualization of the Confusion Matrix.
plt.show()
```
